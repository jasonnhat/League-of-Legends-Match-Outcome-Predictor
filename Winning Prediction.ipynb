```python
import numpy as np
import pandas as pd
import json
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.max_rows', None)
```

```python
# Load data
df = pd.read_csv('games.csv')
champ_data = json.load(open('champion_info.json'))
```

```python
# Add champtions to Data Frame
for i in range(1, 6):
    df[f't1_champ{i}'] = df[f't1_champ{i}id'].apply(lambda x: champ_data['data'][str(x)]['name'])
    df[f't2_champ{i}'] = df[f't2_champ{i}id'].apply(lambda x: champ_data['data'][str(x)]['name'])
```

```python
# Filter columns
df = df[['t1_champ1','t1_champ2','t1_champ3','t1_champ4','t1_champ5','t2_champ1','t2_champ2','t2_champ3','t2_champ4','t2_champ5','firstBlood','firstTower', 'firstBaron', 'firstDragon', 'firstRiftHerald','winner']]
```

```python
# One-hot encode
encodings1 = [pd.get_dummies(df[col], prefix='t1') for col in ['t1_champ1','t1_champ2','t1_champ3','t1_champ4','t1_champ5']]
combined_df1 = sum(encodings1)
encodings2 = [pd.get_dummies(df[col], prefix='t2') for col in ['t2_champ1','t2_champ2','t2_champ3','t2_champ4','t2_champ5']]
combined_df2 = sum(encodings2)

df = df.join(combined_df1).join(combined_df2)
df = df.drop(columns=['t1_champ1','t1_champ2','t1_champ3','t1_champ4','t1_champ5','t2_champ1','t2_champ2','t2_champ3','t2_champ4','t2_champ5'])
X, y = df.drop('winner', axis=1), df['winner']
```

```python
# Split data to train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
clr = RandomForestClassifier(n_jobs=-1)
clr.fit(X_train, y_train)
```

```python
# input function 
def get_prediction_input():
    t1_champs_input = input("Enter the champions for Team 1: ")
    t2_champs_input = input("Enter the champions for Team 2: ")
    firstBlood = int(input("Which team got the first blood (1: Team 1, 2: Team 2): "))
    firstTower = int(input("Which team got the first tower (1: Team 1, 2: Team 2): "))
    firstBaron = int(input("Which team got the first Baron (1: Team 1, 2: Team 2): "))
    firstDragon = int(input("Which team got the first Dragon (1: Team 1, 2: Team 2): "))
    firstRiftHerald = int(input("Which team got the first Rift Herald (1: Team 1, 2: Team 2): "))

    t1_champs = [champ.strip() for champ in t1_champs_input.split(',')]
    t2_champs = [champ.strip() for champ in t2_champs_input.split(',')]

    input_data = {
        'firstBlood': firstBlood,
        'firstTower': firstTower,
        'firstBaron': firstBaron,
        'firstDragon': firstDragon,
        'firstRiftHerald': firstRiftHerald
    }

    for champ in t1_champs:
        col_name = 't1_' + champ
        if col_name in combined_df1.columns:
            input_data[col_name] = 1
        else:
            print(f"Champion {champ} does not exist in the data.")
            input_data[col_name] = 0

    for champ in t2_champs:
        col_name = 't2_' + champ
        if col_name in combined_df2.columns:
            input_data[col_name] = 1
        else:
            print(f"Champion {champ} does not exist in the data.")
            input_data[col_name] = 0

    # Create Data Frame from input_data
    input_df = pd.DataFrame([input_data])
    input_df = input_df.reindex(columns=X.columns, fill_value=0)

    # Prediction
    prediction = clr.predict(input_df)
    return "Team 1 wins" if prediction[0] == 1 else "Team 2 wins"
```

```python
clr.score(X_test, y_test)
```
